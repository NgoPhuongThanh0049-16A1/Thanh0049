```python
pivot_stocks= stocks.pivot_table('close',index = 'date',columns='symbol')
pivot_stocks
```

```python
pivot_stocks= stocks.pivot_table('close',index = 'date',columns='symbol',aggfunc='sum',margins=True,margins_name='total_volume')
pivot_stocks
```

```python
pivot_stocks
```

```python
sorted_pivot_table = pivot_stocks.sort_values(by='total_volume', ascending=False)
sorted_pivot_table.head()
```
